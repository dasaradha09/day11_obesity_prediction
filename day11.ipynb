```python
categorical_features=['Gender','family_history_with_overweight','FAVC','CAEC','SMOKE','SCC','CALC','MTRANS']
```

```python
from sklearn.preprocessing import LabelEncoder
for i in categorical_features:
  lr=LabelEncoder()
  data[i]=lr.fit_transform(data[i])
```

```python
trained_data=data.iloc[:20758,:]
test_data=data.iloc[20758:,:]
```

```python
lr=LabelEncoder()
trained_data['NObeyesdad']=lr.fit_transform(trained_data['NObeyesdad'])
```

```python
trained_data.drop('id',axis=1,inplace=True)
```

```python
trained_data.head()
```

```python
x=trained_data.drop('NObeyesdad',axis=1)
y=trained_data['NObeyesdad']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,stratify=y)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
```

```python

lgbm=LGBMClassifier()
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1,0.2],
    'max_depth': [5, 10, 15]
}

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=lgbm, param_grid=param_grid, cv=3, scoring='accuracy')

# Fit the model to the training data
grid_search.fit(x_train, y_train)

# Get the best parameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Print the best parameters
print("Best Hyperparameters:", best_params)

# Evaluate the best model on the test set
y_pred = best_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```
